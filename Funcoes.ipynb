```python
import pandas as pd
import os
```

```python
#Função para Abrir os arquivos CSV e transformar em DataFrame
def csv_para_df (pasta):
    arquivos = [l for l in os.listdir(pasta) if l.endswith('.csv')]
    lista_df = []

    for arquivo in arquivos:
        caminho_arquivo = os.path.join(pasta, arquivo)
        try:
            df = pd.read_csv(caminho_arquivo, sep=',', encoding='utf-8')
            lista_df.append(df)
        except Exception as e:
            print(f"Erro ao ler o arquivo {arquivo}: {e}")
    
    df_final = pd.concat(lista_df, ignore_index=True)
    return df_final
```

```python
#Função para Retirar Espaços e Formatar Texto
def limpar_textos(df, lista_colunas):
    for col in lista_colunas:
        if col in df.columns:
            print(f'Limpando a Coluna {col}')
            df[col] = df[col].astype(str).str.strip().str.upper() # Letras em Maiúsculo
            df[col] = df[col].astype(str).str.title() # Letras em Formato de Título

    return df
```

```python
#Função para Transformar Colunas em Datas
def transformar_datas(df, lista_colunas):
    for col in lista_colunas:
        if col in df.columns:
            print(f'Transformando a Coluna {col} em Data')
            df[col] = pd.to_datetime(df[col], errors='coerce', yearfirst=True) # Transformar em Data (AAAA/MM/DD)
    return df
```

```python
#Ordenação Decrescente
def ordernar_desc(df, nome_coluna):
    linhas = df.to_dict(orient='records') # Converte DataFrame em lista de dicionários
    contador = 0
    for i in range(1, len(linhas)):
        chave = linhas[i] 
        j = i - 1 # Índice do elemento anterior
        while j >= 0 and linhas[j][nome_coluna] < chave[nome_coluna]: # Comparação para ordem decrescente
            linhas[j + 1] = linhas[j] 
            j -= 1
            contador += 1
        linhas[j + 1] = chave # Insere a chave na posição correta
    print(f'Quantidade de Comparações: {contador}')
    return pd.DataFrame(linhas)
```
